1. **Export it as a GraphML file** named for instance 'INFOMEDIA NER PMI NETWORK LAYOUT.graphml'. Put it in the same folder as this notebook.
1. Edit the settings below
1. Run all the cells from the menu: Cell > Run all.
1. It's over when the last cell outputs "Done."
